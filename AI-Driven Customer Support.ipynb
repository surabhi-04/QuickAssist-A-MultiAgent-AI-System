```python
import pandas as pd
from crewai import Agent, Task, Crew, Process
import re
import string
from typing import List, Dict
import os
from langchain.chat_models import ChatOpenAI
import time

openai_api_key = "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"
os.environ["OPENAI_API_KEY"]=openai_api_key
os.environ["OPENAI_MODEL_NAME"] = 'gpt-3.5-turbo'

# 1. Load and Preprocess Data
try:
    historical_data = pd.read_csv("Historical_ticket_data.csv")
except FileNotFoundError:
    print("Error: Historical_data_set.CSV not found. Please upload it to Colab or provide the correct path.")
    historical_data = pd.DataFrame()

def clean_text(text: str) -> str:
    """Cleans text by removing HTML, URLs, special chars, and extra spaces."""
    if not isinstance(text, str):
        return ""
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^\w\s' + string.punctuation + ']', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def preprocess_historical_data(historical_data: pd.DataFrame) -> pd.DataFrame:
    """Preprocesses historical data, cleaning text and handling missing values."""
    if historical_data.empty:
        return pd.DataFrame()
    for col in ['Customer Query', 'Resolution Steps']:
        if col in historical_data.columns:  # Check if the column exists
            historical_data[col] = historical_data[col].apply(clean_text)
        else:
            historical_data[col] = "" # create empty column
    for col in ['Category', 'Conversation ID']:
        if col in historical_data.columns:
             historical_data[col].fillna('Unknown', inplace=True)
        else:
             historical_data[col] = "Unknown"
    return historical_data

if not historical_data.empty:
    historical_data = preprocess_historical_data(historical_data)

# 2. Define Tools (Simplified for CrewAI -  Direct function calls)
def analyze_conversation(conversation: str) -> Dict:
    """Analyzes conversation for summary, sentiment, priority, actions, category, intent."""
    # Placeholder implementation using string matching and simple logic
    #  Replace with actual NLP models (e.g., Hugging Face Transformers) for better results
    summary = conversation[:50] + "..." if len(conversation) > 50 else conversation
    sentiment = "Positive" if "thanks" in conversation.lower() else "Neutral"
    priority = "High" if "urgent" in conversation.lower() or "critical" in conversation.lower() else "Medium" if "frustrated" in conversation.lower() or "annoyed" in conversation.lower() else "Low"
    actions = [
        "reset" if "reset" in conversation.lower() else None,
        "rollback" if "roll back" in conversation.lower() else None,
        "discount" if "discount" in conversation.lower() else None,
        "check" if "check" in conversation.lower() else None,
        "clear" if "clear" in conversation.lower() else None,
        "upgrade" if "upgrade" in conversation.lower() else None,
        "disable" if "disable" in conversation.lower() else None,
        "send" if "send" in conversation.lower() else None,
        "install" if "install" in conversation.lower() else None,
    ]
    actions = [action for action in actions if action is not None]
    category = "Technical Support"  # Default
    intent = "Problem Report"  # Default
    return {
        "summary": summary,
        "sentiment": sentiment,
        "priority": priority,
        "actions": actions,
        "category": category,
        "intent": intent,
    }

def recommend_resolution(analysis_output: Dict, historical_data: pd.DataFrame) -> str:
    """Recommends resolution based on historical data."""
    if historical_data.empty:
        return "No historical data available."

    query = analysis_output["summary"]
    if not query:
        return "Cannot provide recommendation without a query."

    # Placeholder:  Simplified similarity search (keyword-based)
    best_match = {"index": -1, "similarity": 0}
    for index, row in historical_data.iterrows():
        similarity = sum(1 for word in query.lower().split() if word in row['Customer Query'].lower())
        if similarity > best_match["similarity"]:
            best_match["similarity"] = similarity
            best_match["index"] = index

    if best_match["index"] != -1 and best_match["similarity"] > 0:
        conversation_id = historical_data.iloc[best_match["index"]]['Conversation ID']
        resolution = historical_data.iloc[best_match["index"]]['Resolution Steps']
        return f"Recommended Resolution (from Conversation ID: {conversation_id}): {resolution}"
    else:
        return "No similar historical issue found."

def route_task(analysis_output: Dict) -> str:
    """Routes task to appropriate team."""
    priority = analysis_output["priority"]
    actions = analysis_output["actions"]
    category = analysis_output["category"]

    if priority == "High":
        return "Urgent Response Team"
    elif "escalate" in actions:
        return "Escalation Team"
    elif category == "Billing":
        return "Billing Department"
    elif category == "Sales":
        return "Sales Team"
    elif category == "Account Management":
        return "Account Management Team"
    elif category == "Technical Support":
        return "Technical Support Team"
    else:
        return "General Support Team"

def estimate_resolution_time(analysis_output: Dict, historical_data: pd.DataFrame) -> str:
    """Estimates resolution time based on historical data."""
    if historical_data.empty:
        return "Resolution time estimation unavailable."

    category = analysis_output["category"]
    matching_data = historical_data[historical_data['Category'] == category]
    if not matching_data.empty and 'Resolution Time' in matching_data.columns:
        avg_time = matching_data['Resolution Time'].mean()
        return f"Estimated Resolution Time: {avg_time:.2f} hours"
    else:
        return "Resolution time estimation unavailable for this category."

# 3. Define Agents
conversation_analyzer_agent = Agent(
    name="Conversation Analyzer",
    role="Analyzes customer conversations to understand the issue.",
    goal="Extract key information from customer interactions.",
    backstory="A highly skilled AI agent with expertise in natural language processing and customer support analysis.",
    # No tools, will call function directly.  This is simplified for CrewAI.
    allow_delegation=False
)

resolution_recommender_agent = Agent(
    name="Resolution Recommender",
    role="Recommends solutions based on historical data.",
    goal="Provide accurate and efficient resolutions to customer issues.",
    backstory="An AI agent with access to a vast database of past customer interactions and resolutions.",
    allow_delegation=False
)

task_router_agent = Agent(
    name="Task Router",
    role="Routes tasks to the appropriate team.",
    goal="Ensure that customer issues are handled by the right team or agent.",
    backstory="An AI agent with knowledge of team structures and expertise in task assignment.",
    allow_delegation=False
)

time_estimator_agent = Agent(
    name="Resolution Time Estimator",
    role="Estimates the time required to resolve an issue.",
    goal="Provide accurate time estimates to manage customer expectations and optimize workflow.",
    backstory="An AI agent trained on historical resolution data and experienced in predicting resolution times.",
    allow_delegation=False
)

orchestrator_agent = Agent( #added orchestrator agent
    name="Orchestrator",
    role="Orchestrates the workflow of the agents",
    goal="Manage the process",
    backstory="This agent manages the workflow",
    allow_delegation=True
)

# 4. Define Tasks
#Instead of defining tasks, we will execute the functions directly within a main function

# 5. Create Crew
crew = Crew(
    agents=[orchestrator_agent, conversation_analyzer_agent, resolution_recommender_agent, task_router_agent, time_estimator_agent],
    tasks=[], # tasks will be created and assigned dynamically
    process=Process.sequential  # Can be changed to hierarchical if needed
)


def run_customer_support_crew(conversation: str, historical_data: pd.DataFrame) -> Dict:
    """Runs the customer support crew for a given conversation."""
    # Use orchestrator agent to manage the flow

    # Create a Task object for analyze_conversation
    # Added expected_output with the type hint for the output of analyze_conversation
    analysis_task = Task(
        name="Analyze Conversation",
        description="Analyze the customer conversation to extract key information.",
        func=lambda : analyze_conversation(conversation), # Assign the lambda function to the 'func' attribute
        expected_output="Dict" # Add expected_output to specify the expected type of the output
    )
    analysis_output = conversation_analyzer_agent.execute_task(analysis_task) # Pass the Task object to execute_task

    #print(f"Conversation Analysis: {analysis_output}") #for debugging

    # Create a Task object for recommend_resolution
    # Added expected_output with the type hint for the output of recommend_resolution
    recommendation_task = Task(
        name="Recommend Resolution",
        description="Recommend a resolution based on the analysis output and historical data.",
        func=lambda : recommend_resolution(analysis_output, historical_data),
        expected_output="str" # Add expected_output to specify the expected type of the output
    )
    recommendation = resolution_recommender_agent.execute_task(recommendation_task)
    #print(f"Recommendation: {recommendation}")

    # Create a Task object for route_task
    # Added expected_output with the type hint for the output of route_task
    route_task_obj = Task(
        name="Route Task",
        description="Route the task to the appropriate team based on analysis output.",
        func=lambda : route_task(analysis_output),
        expected_output="str" # Add expected_output to specify the expected type of the output
    )
    route = task_router_agent.execute_task(route_task_obj)

    #print(f"Route: {route}")
    # Create a Task object for estimate_resolution_time
    # Added expected_output with the type hint for the output of estimate_resolution_time
    estimation_task = Task(
        name="Estimate Resolution Time",
        description="Estimate the resolution time based on analysis output and historical data.",
        func=lambda : estimate_resolution_time(analysis_output, historical_data),
        expected_output="str" # Add expected_output to specify the expected type of the output
    )
    estimation = time_estimator_agent.execute_task(estimation_task)

    return {
        "analysis": analysis_output,
        "recommendation": recommendation,
        "route": route,
        "estimation": estimation,
    }






if __name__ == "__main__":
    conversations = [
        """Hello! My project data isn’t syncing between my laptop and tablet.
        Changes on one device don’t show up on the other. Can you help?""",
        """Hi, I’m really frustrated. Your smart home app crashes every time I
        try to connect my older thermostat model. It worked fine on my old phone!""",
        """Good morning! I’m having an issue where my app keeps saying ‘no internet
        connection,’ but my Wi-Fi is working fine. Other apps load normally.""",
        """Hi, this is urgent! Your API is rejecting our payment gateway
        integration. Error: ‘Invalid SSL certificate.’ Our cert is valid
        and up-to-date!""",
        """Hi there! I’ve been trying to install the latest update for your design
        software for hours. It keeps failing at 75% with an ‘unknown error.’
        What’s wrong?""",
        "I can't log in to my account.",  # test case
        "How do I reset my password?",  # test case
        "The app is not working",  # test case
    ]

    for i, conv in enumerate(conversations):
        print(f"\n--- Conversation {i + 1} ---")
        print(f"Conversation: {conv}")
        results = run_customer_support_crew(conv, historical_data)
        print(f"Analysis: {results['analysis']}")
        print(f"Recommendation: {results['recommendation']}")
        print(f"Route: {results['route']}")
        print(f"Estimated Resolution Time: {results['estimation']}")
        time.sleep(1)
```
